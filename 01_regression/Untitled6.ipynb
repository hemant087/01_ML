```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x = 4*np.random.rand(300,1)-3
```

```python
# y = 2*x**2 + 0.2*x+3
y = 2*x**2 + 0.2*x+10+ np.random.rand(300,1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
```

```python
x_train_transform = poly.fit_transform(x_train)
```

```python
x_test_transform = poly.fit_transform(x_test)
```

```python
x_test_transform = x_test_transform.reshape(-1,1)
```

```python
# Transform input to polynomial features
degree = 2  # You can change degree here
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(x)

# Fit the model
model = LinearRegression()
model.fit(X_poly, y)

# Predict
y_pred = model.predict(X_poly)

# Plot
plt.scatter(x, y, color='blue', label='Original Data')
plt.scatter(x, y_pred, color='red', label=f'Polynomial Regression (degree={degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Polynomial Regression Example')
plt.grid(True)
plt.show()
```
